```python
import numpy as np
from netCDF4 import Dataset
import wrf
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon as Poly
```

```python
nc = Dataset('../../wrfout_d01_2018-05-07_18:00:00')
```

```python
ppn = wrf.getvar(nc, 'RAINNC', timeidx=wrf.ALL_TIMES)
```

```python
ppn.dims
```

```python
ppn.where(ppn.south_north + ppn.west_east < 100, drop=True).mean()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
bm = wrf.get_basemap(ppn)

lats, lons = wrf.latlon_coords(ppn)
x, y = bm(wrf.to_np(lons), wrf.to_np(lats))

bm.readshapefile('../wrfplot/shapefiles/provincias', 'prov', drawbounds = False)

patches   = []

for info, shape in zip(bm.prov_info, bm.prov):
    if info['NAM'] == 'CÓRDOBA':
        patches.append( Polygon(np.array(shape), True) )
        p = Poly(shape)
print patches

ax.add_collection(PatchCollection(patches, facecolor= 'm', edgecolor='k', linewidths=1., zorder=2))
bm.contourf(x, y, wrf.to_np(ppn[10]), 10, cmap="Blues")

plt.show()
```
